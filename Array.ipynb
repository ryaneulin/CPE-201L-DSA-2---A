```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")
def reverse_arr(arr):
    print("Reversing all Elements")
    reversed_arr = arr[::-1]
    for i, element in enumerate(reversed_arr):
        print(f"Index {i}: {element}")

def delete_array(arr):
    print("""
            1. Delete All
            2. Delete Only One
            3. Exit
          """)

    choice = int(input("Pick 1 - 3: "))

    if choice == 1:
        arr.clear()
        print(arr)

    elif choice == 2:
        for i, element in enumerate(arr):
            print(f"Index {i}, {element}")

        ask = int(input("What Index do you want to delete: "))
        arr.pop(ask)
        for new, array in enumerate(arr):
            print(f"Index {new}, {array}")

    else:
        print("Back to the Main Menu")


def sort_array(arr):
    print("Sorting all Elements")
    sorted_arr = sorted(arr)
    for i, element in enumerate(sorted_arr):
        print(f"Index {i}, {element}")

def modify_array(arr):
    print("Modifying all Elements")
    num_append = int(input("What number do you want to append: "))
    arr.append(num_append)
    for i, element in enumerate(arr):
        print(f"Index {i}, {element}")

def main():
    # Initialize an array
    arr = [40, 20, 50, 30, 10]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reversing array")
        print("3. Delete_array")
        print("4. Sort_array")
        print("5. Modify_array")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")
        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse_arr(arr)
        elif choice == '3':
            delete_array(arr)
        elif choice == '4':
            sort_array(arr)
        elif choice == '5':
            modify_array(arr)
        else:
            print("\nThankyou")
            break

if __name__ == "__main__":
    main()
```
